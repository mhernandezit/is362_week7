# 1. Load the ratings by user information that you collected into a pandas dataframe.


```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

movie_data = pd.read_csv('movie_data.csv', index_col=0)
movie_data
```

# 2. Show the average ratings for each user and each movie.
First, we build out the User Average column by performing the mean function on the x axis.

```python
movie_data['User Average'] = movie_data.mean(axis = 1, skipna = True)
movie_data
```

Next we set the Average Movie Rating by performing the mean function on the y axis.

```python
movie_data.loc['Average Movie Rating'] = movie_data.mean(axis = 0, skipna = True)
movie_data
```

# 3. Create a new pandas dataframe, with normalized ratings for each user. Again, show the average ratings for each user and each movie.

For normalizing the data, we first transform the data using the min/max scaler imported from skikit learn.  The normalized dataframe does not have any column or row headers, so the rest of the work done here is adding back that information.

```python
normalized_data = movie_data.copy()
min_max_scaler = preprocessing.MinMaxScaler()
scaled = min_max_scaler.fit_transform(normalized_data)
movie_normalized = pd.DataFrame(scaled)
movie_normalized.columns = ['Captain Marvel', 'Aquaman', 'Us','Green Book','The Lego Movie 2','Batman Begins','User Average']
names = ['Mike','Linda','Ray','Liz','Gabriel','Average Movie Rating']
movie_index = pd.DataFrame(names, columns=['a'])
movie_normalized.insert(0,'Names',movie_index['a'])
movie_normalized.set_index(['Names'])
```

# 4. Provide a text-based conclusion: explain what might be advantages and disadvantages of using normalized ratings instead of the actual ratings.

Normalized ratings are good to show if some movies either rate similarly among people, or if there are a variety of different ratings available.  With a larger number of reviews normalized ratings would be able to tell if a movie has ratings above a certain threshold.  This is useful when movie ratings are similar between movies.  At the same point, it may be better to see the actual rating when deciding on a movie to watch - the mean in this case has more meaning that the normalized rating.
